```python


import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

# Définition des paramètres
duration = 0.002  # durée de la forme d'onde en secondes
sampling_rate = 200e6  # fréquence d'échantillonnage en Hz
pulse_width = 0.000075  # largeur d'impulsion en secondes
pulse_amplitude = 4.0  # amplitude de l'impulsion
noise_amplitude = 0.1  # amplitude du bruit blanc gaussien
freq_max = 1420  # MHz
freq_bw = 1400  # MHz
pulse_pa = 42

# Calcul du nombre d'échantillons
num_samples = int(duration * sampling_rate)

# Génération d'un signal de bruit blanc gaussien complexe
noise = np.random.normal(0, noise_amplitude, num_samples) + 1j * np.random.normal(0, noise_amplitude, num_samples)

# Création d'un vecteur temporel échantillonné à une fréquence d'échantillonnage de sampling_rate Hz
t = np.arange(num_samples) / sampling_rate

# Détermination de la phase de la polarisation linéaire
pa_rad = np.deg2rad(pulse_pa)
phase = 2*np.pi*np.random.rand(num_samples)
phase_signal = np.zeros(num_samples)
phase_signal[int(num_samples/2 - 3*pulse_width/2*sampling_rate) : int(num_samples/2 + 3*pulse_width/2*sampling_rate)] = pa_rad

phase = phase + phase_signal

# Initialisation du vecteur de pulsation avec polarisation linéaire
pulse_with_polarization = pulse_amplitude * np.exp(1j * phase)

# Application d'une fenêtre gaussienne à la pulsation
pulse = signal.windows.gaussian(num_samples, std=sampling_rate * pulse_width) * pulse_with_polarization

# Ajout du signal pulsé au bruit blanc gaussien complexe
signal = noise + pulse

# Normalisation de la puissance du signal à 1
signal /= np.sqrt(np.mean(np.abs(signal)**2))

def plot_raw(wavevforme):
    # Calcul de la norme du signal
    signal_norm = np.abs(wavevforme)

    # Affichage du signal généré et de sa norme
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))
    ax1.plot(t, np.real(wavevforme), label='Partie réelle')
    ax1.plot(t, np.imag(wavevforme), label='Partie imaginaire')
    ax1.set_xlabel('Temps (s)')
    ax1.set_ylabel('Voltage (V)')
    ax1.legend()

    ax2.plot(t, signal_norm, label='Norme')
    ax2.set_xlabel('Temps (s)')
    ax2.set_ylabel('Voltage (V)')
    ax2.legend()

    plt.show()
    
plot_raw(signal)
```

```python

def fft(raw, fftlen):
        # print("fft")
        nchan, nschan_perpol, npol = np.shape(raw)
        nfft = nschan_perpol / fftlen
        # print('0:',np.shape(raw.wav))
        raw.shape = (nchan, int(nfft), fftlen, npol)
        raw = np.reshape(raw, (nchan, int(nfft), fftlen, npol))
        print('1:',np.shape(raw))
        raw = np.fft.fft(raw, axis=2)
        #raw = np.fft.fftshift(raw, axes=2) #ifftshift
        print('2:',np.shape(raw))
        return raw

def ifft(raw):
        # print("ifft")
        nchan, nschan_perpol, fftlen, npol = np.shape(raw)
        # raw.wav = np.fft.ifftshift(raw.wav, axes=2)
        raw = np.fft.ifft(raw, axis=2)
        # raw.wav.shape = (nchan, nschan_perpol*fftlen, npol)
        raw = np.reshape(raw, (nchan, nschan_perpol * fftlen, npol))
        # print('3:',np.shape(raw.wav))
        return raw

def coherent_dedisp(raw, dm=0, freq=freq_max, chan_bw=freq_bw):
        print("coherent_dedisp")
        # raw.dm = 2
        nschan_file = len(raw)
        nof_polcpx = 1
        if(dm != 0):
            dmfac = dm * 2.0 * np.pi / 2.41e-10  # (2.41e-10*(1.0+s->earth_z4/1.0e4)) doppler TODO
            nchan, max_fftlen, npol = np.shape(raw)
            raw = fft(raw, fftlen=max_fftlen)
            chirp = np.zeros(max_fftlen) + 1j * np.zeros(max_fftlen)
            for ichan in range(nchan):
                max_fftlen_vec = np.linspace(0, max_fftlen - 1, max_fftlen)
                # for i in range(max_fftlen):
                dfreq = max_fftlen_vec * chan_bw / max_fftlen
                dfreq[(max_fftlen_vec > max_fftlen / 2 - 1)] -= chan_bw  # if not use fftshift
                freqfac = dfreq / freq
                freqfac = freqfac * freqfac / (freq + dfreq)
                arg = dmfac * freqfac
                chirp[:].real = (np.cos(arg) / max_fftlen)
                chirp[:].imag = -1.0 * (np.sin(arg) / max_fftlen)
                for ipol in range(npol):
                    wav = np.copy(raw[ichan, :, :, ipol]) * max_fftlen / nschan_file / nof_polcpx
                    raw[ichan, :, :, ipol].real = wav.real * chirp.real - wav.imag * chirp.imag
                    raw[ichan, :, :, ipol].imag = wav.imag * chirp.real + wav.real * chirp.imag
            raw = ifft(raw)
        else:
            print("WARNING: no need to coherently dedisperse at DM=0")
        return raw
```

```python



nchan = 1
max_fftlen = len(signal)
npol = 1
signal_reshape = np.reshape(np.copy(signal), (nchan, max_fftlen, npol))
print('signal_reshape', np.shape(signal_reshape))
signal_disp = coherent_dedisp(signal_reshape, dm=-1e-2, freq=freq_max, chan_bw=freq_bw)
plot_raw(np.squeeze(signal_disp))
print('signal_disp', np.shape(signal_disp))
signal_fft = fft(signal_disp, fftlen=1000)
print('signal_fft', np.shape(signal_fft))
signal_fft = np.fft.fftshift(signal_fft, axes=2)

def plot_tf(signal_tmp, dt=5): 
    print(np.shape(signal_tmp))
    nonne, max_fftlen, nchan, npol = np.shape(signal_tmp)
    signal_tmp = np.reshape(signal_tmp, (int(max_fftlen/dt), int(dt), int(nchan)))
    signal_tmp = np.mean(signal_tmp, axis=1)
    signal_tmp = np.reshape(signal_tmp, (int(max_fftlen/dt), int(nchan)))
    print(np.shape(signal_tmp))
    sq = np.abs(signal_tmp)
    plt.imshow(np.rot90(sq), interpolation='none', cmap='afmhot', aspect='auto', extent=[
                   0, t[-1], freq_max - freq_bw, freq_max])
    plt.ylabel('Freq (MHz)')
    plt.xlabel('Temps (s)')
    plt.show()

    
plot_tf(signal_fft)
#plot_raw(np.squeeze(signal_fft))
```
